## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
______________________________________________________________________________________________There are four basic types of SQL joins: inner, left, right, and full.Inner Join gets all records that match the condition in both tables, and records in both tables that do not match are not reported. Left Join returns all the rows from the left table in conjunction with the matching rows from the right table. Right Join returns all the rows from the right table in conjunction with the matching rows from the left table. Outter Join combines left join and right join. It returns rows from either table when the conditions are met and returns NULL value when there is no match.
